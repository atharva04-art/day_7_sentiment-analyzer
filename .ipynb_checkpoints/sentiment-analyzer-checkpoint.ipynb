```python
from transformers import pipeline 

#  Load the sentiment analysis model 
sentiment_model = pipeline("sentiment-analysis")

#  Test it
sentiment_model("This weather is absolutely amazing!")
```

```python
sentences = [
    "I love this weather, it's so calm and peaceful!", 
    "this is the worst experinece I've had.",
    "I feel okay today.",
    "I am not sure how to feel about this.",
    "Absolutely fantastic movie!"
]

#  Run prediction
for sentence in sentences:
    result = sentiment_model(sentence)[0]
    print(f"Text: {sentence}")
    print(f"sentiment: {result['label']}, score: {result['score']:.2f}")
    print("-" * 50)
```

```python
import matplotlib.pyplot as plt

results = sentiment_model(sentences)

labels = [res['label'] for res in results]
scores = [res['score'] for res in results]

#  Plotting
plt.figure(figsize=(8, 4))
bars = plt.bar(sentences, scores, color=['green' if label == 'positive' else 'red' for label in labels])
plt.xticks(rotation=45, ha='right')
plt.ylabel("Confidence Score")
plt.title("Sentiment Score")

#  Add score labels on top of bar 
for bar, score in zip(bars, scores):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval + 0.01, f"{score:.2f}", ha='center', va='bottom')


plt.tight_layout()
plt.show()
#  Optional: clear after plot (if looping or repeating)
plt.clf()
plt.close('all')
```
